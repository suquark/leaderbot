```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 4
A = np.zeros((n+1, n+1), dtype=float)

for i in range(0, n+1):
    for j in range(0, n+1):
        for k in range(0, n+1):
            A[i, j] += (-1)**k * scipy.special.binom(j, k) * scipy.special.binom(n - j, i - k)
```

```python
p = plt.imshow(A)
plt.colorbar(p)
```

```python
# Chebyshev
n = 50
A = np.zeros((n, n), dtype=float)

for i in range(1, n+1):
    for j in range(1, n+1):
        A[i-1, j-1] = np.sqrt(2/n) * np.cos((np.pi / n) * i*j + np.pi*j)
```

```python
# DCT-IV
n = 50
A = np.zeros((n, n), dtype=float)

for i in range(1, n+1):
    for j in range(1, n+1):
        A[i-1, j-1] = np.sqrt(2/n) * np.cos((np.pi / n) * (i-0.5)*(j-0.5))
```

```python
# 
n = 4
A = np.zeros((n, n), dtype=float)

for i in range(1, n+1):
    for j in range(1, n+1):
        A[i-1, j-1] = np.cos((np.pi / n) * (i)*(j) + n*np.pi)
```

```python
A.T @ A
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(11, 4))
p0 = ax[0].imshow(A.T@A)
p1 = ax[1].imshow(A)
plt.colorbar(p0)
plt.colorbar(p1)
```
