```python
import texplot
import pickle
import leaderbot as lb

filename = '../benchmark/models_train_full.pkl'
# filename = '../benchmark/models_train_full_both_ties.pkl'
with open(filename, 'rb') as f:
    results = pickle.load(f)

models = results['models']
test_data = results['test_data']

model = models[22]  # Davidson Scaled (k conv = 0 and k tie = 0)
```

```python
import matplotlib.pyplot as plt
latex = True
save = True
```

```python
%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}  # to avoid cutting off z label
#%matplotlib tk
#%matplotlib inline
fig, ax = plt.subplots(figsize=(8, 6.5), subplot_kw={'projection': '3d'},
                       gridspec_kw=dict(top=1.28, bottom=0.06, left=-0.06, right=0.895))

model.map_distance(ax, max_rank=50, method='kpca', dim='3d', sign=(1, -1, -1),
                   latex=latex, save=save)
```

```python
with texplot.theme(use_latex=latex):
    fig, ax = plt.subplots(figsize=(12, 5.25))
    ax.axis('equal')
    plt.subplots_adjust(left=0.05, right=0.945, top=0.95, bottom=0.1)

    model.map_distance(ax, max_rank=50, method='mds', dim=(2, 0), sign=(-1, -1),
                       latex=latex, save=save)
    plt.tight_layout()
```
