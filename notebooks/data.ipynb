```python
import requests
import json

url = "https://storage.googleapis.com/arena_external_data/public/clean_battle_20240814_public.json"

# Get the content from the URL
response = requests.get(url, stream=True)

# Raise an error if the request was not successful
response.raise_for_status()

# Save the content to a local file
with open("clean_battle_20240814_public.json", "wb") as f:
    for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)

print("File downloaded and saved as 'clean_battle_20240814_public.json'.")
```

```python
with open("clean_battle_20240814_public.json", "r") as f:
    # Load only the first few kilobytes to inspect the structure
    for i, line in enumerate(f):
        if i > 100:  # Stop after reading 100 lines
            break
        print(line)
```
