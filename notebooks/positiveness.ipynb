```python
import numpy as np
```

```python
# Test tii + tjj - 2 tij is positive on some random data
# Prove using permutationt to bring i j to the 2x2 principal sub-matrix,
# and use Cholesky decomposition. basically, any principal submatrix is PD.

n, m = 10, 3
A = np.random.randn(n, m)
A = A @ A.T

for i in range(n-1):
    for j in range(i+1, n):
        B = np.array([[A[i, i], A[i, j]], [A[j, i], A[j, j]]])
        e = np.linalg.eigh(B)[0]
        s = B[0, 0] + B[1, 1] - 2.0 * B[0, 1]
        print(f'{e[0]:>9.6f}, {e[0]:>9.6f} | {s:>9.6f}')
```
